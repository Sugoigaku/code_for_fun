# 计算学生模拟考平均分，理解度平均值，作业正确率

   * 2019冬季作业登记表

```python
import  pandas as pd
import numpy as np 
import os 
from pandas import DataFrame,Series
import re
import xlrd
```

## 2.计算思路

* 模拟考平均分  

```python
df = pd.read_excel("2019冬季.xlsx",sheetname="情报",header=1)
test_score_df = df[['姓名','总考分1','总考分2','总考分3']]
test_score_df['考试平均分']=test_score_df.mean(axis=1)
test_score_df.head()
```

```python
test_score_df.shape
```

* 作业正确率

> 合并2019冬季作业登记表当中的情报各班表格

```python
file = '2019冬季作业登记表.xlsx'

# 处理数据，合并多个sheet
wb = xlrd.open_workbook(file)
# 获取全部的工作簿
sheets = wb.sheets()
qingbao_df = DataFrame()
# 从第4个sheet开始合并,到第7个
for i in range(4,7):
    # header=1，那么数据会从第二行开始读取
    temp_df = pd.read_excel(file, sheet_name=i, header=1)
    # how='all'：只要存在数据为空的行都删除，还有其他的方法
    # 每个sheet倒数两行都不要
    dropna_df = temp_df.dropna(how='all')[:-2]
    qingbao_df = qingbao_df.append(dropna_df)
qingbao_df.head()
```

> 取得算平均的区域

```python
qingbao_score = qingbao_df.ix[:,[i for i in range(1,10)]]
qingbao_score.head()
```

> 计算专业课作业平均正确率

```python
qingbao_df["CS_Average"]=qingbao_score.mean(axis=1)
qingbao_average = qingbao_df[['姓名','CS_Average']]
qingbao_average.head()
```

> 计算数学课课作业平均正确率与上面同理

```python
file = '2019冬季作业登记表.xlsx'

# # 处理数据，合并多个sheet
# wb = xlrd.open_workbook(file)
# # 获取全部的工作簿
# sheets = wb.sheets()

shuxue_df = DataFrame()
# 从第10个sheet开始合并
for i in range(4):
    # header=1，那么数据会从第二行开始读取
    temp_df = pd.read_excel(file, sheet_name=i, header=1,index=False)
    # how='all'：只要存在数据为空的行都删除，还有其他的方法
    # 每个sheet倒数两行都不要
    dropna_df = temp_df.dropna(how='all')[:-2]
    shuxue_df = shuxue_df.append(dropna_df)
shuxue_df.head()
```

```python
shuxue_score = shuxue_df.ix[:,[i for i in range(2,15)]]
shuxue_df['Math_Average']=shuxue_score.mean(axis=1)
shuxue_average = shuxue_df[['姓名','Math_Average']]
shuxue_average.head()
```

> 专业课作业成绩 left join 数学作业成绩 得出情报专业同学的总作业准确率，再求平均 

```python
homework = pd.merge(qingbao_average, shuxue_average, how='left', on=['姓名'])
homework['作业平均正确率'] = homework.mean(axis=1)
homework.head()
```

```python
homework.shape
```

* 理解度计算

> 专业课理解度计算 情报A

```python
def c_average(file, dataframe, sheets_num):
    for i in range(1,sheets_num):
        temp_df = pd.read_excel(file, sheet_name=i,index=False)
        temp_df = temp_df.ix[:,[i for i in range(1,len(temp_df)-2)]]
        dropna_df = temp_df.dropna(how='all')
        dataframe = pd.merge(dataframe, dropna_df, how='outer', on=['姓名'])
    return dataframe
```

```python
file = '情报A（回复）.xlsx'
a_df = DataFrame()

wb = xlrd.open_workbook(file)
# 获取全部的工作簿
sheets = wb.sheets()

temp_df = pd.read_excel(file, sheet_name=0,index=False)
temp_df = temp_df.ix[:,[i for i in range(1,3)]]
dropna_df = temp_df.dropna(how='all')
a_df = a_df.append(dropna_df)

for i in range(1,len(sheets)):
    temp_df = pd.read_excel(file, sheet_name=i,index=False)
    temp_df = temp_df.ix[:,[i for i in range(1,3)]]
    temp_df.drop_duplicates(['姓名'],keep='last')
    dropna_df = temp_df.dropna(how='all')
    a_df = pd.merge(a_df, dropna_df, how='outer', on=['姓名'])
    
a_df['平均A'] = a_df.mean(axis=1)
a_df = a_df[['姓名','平均A']]
a_df = a_df.groupby(['姓名']).max()
a_df.shape
```

> 专业课理解度计算 情报B

```python
file = '情报B（回复）.xlsx'
b_df = DataFrame()

wb = xlrd.open_workbook(file)
# 获取全部的工作簿
sheets = wb.sheets()

temp_df = pd.read_excel(file, sheet_name=0,index=False)
len_colum = temp_df.shape[1]
temp_df = temp_df.ix[:,[i for i in range(1,len_colum-3)]]
dropna_df = temp_df.dropna(how='all')
b_df = b_df.append(dropna_df)

for i in range(1,len(sheets)):
    temp_df = pd.read_excel(file, sheet_name=i,index=False)
    col_len = temp_df.shape[1]
    temp_df = temp_df.ix[:,[i for i in range(1,col_len-3)]]
    dropna_df = temp_df.dropna(how='all')
    b_df = pd.merge(b_df, dropna_df, how='outer', on=['姓名'])
# b_df = c_average(file,b_df,len(sheets))

b_df['平均B'] = b_df.mean(axis=1)
b_df = b_df[['姓名','平均B']]
b_df = b_df.groupby(['姓名']).max()
b_df.shape
```

> 专业课理解度计算 情报C

```python
file = '情报C（回复）.xlsx'
c_df = DataFrame()

wb = xlrd.open_workbook(file)
# 获取全部的工作簿
sheets = wb.sheets()

temp_df = pd.read_excel(file, sheet_name=0,index=False)
len_colum = temp_df.shape[1]
temp_df = temp_df.ix[:,[i for i in range(1,len_colum-3)]]
dropna_df = temp_df.dropna(how='all')
c_df = c_df.append(dropna_df)

for i in range(1,len(sheets)):
    temp_df = pd.read_excel(file, sheet_name=i,index=False)
    col_len = temp_df.shape[1]
    temp_df = temp_df.ix[:,[i for i in range(1,col_len-3)]]
    dropna_df = temp_df.dropna(how='all')
    c_df = pd.merge(c_df, dropna_df, how='outer', on=['姓名'])
# b_df = c_average(file,b_df,len(sheets))

c_df['平均C'] = c_df.mean(axis=1)
c_df = c_df[['姓名','平均C']]
c_df = c_df.groupby(['姓名']).max()
c_df.head()
```

> 专业课总平均

```python
zhuanmen_comp = DataFrame()
zhuanmen_comp = pd.merge(a_df, b_df, how='outer', on=['姓名'])
zhuanmen_comp = pd.merge(zhuanmen_comp, c_df, how='outer', on=['姓名'])
```

```python
zhuanmen_comp['专业平均'] = zhuanmen_comp.mean(axis=1)
zhuanmen_comp.head()
```

> 数学情报班理解度

```python
data = {'姓名':[]}
shuxue_comp = DataFrame(data)
for ch in 'ABCD':
    file = '数学{}（回复）.xlsx'.format(ch)
    print(file)
    d_df = DataFrame()
    wb = xlrd.open_workbook(file)
    # 获取全部的工作簿
    sheets = wb.sheets()
    temp_df = pd.read_excel(file, sheet_name=0,index=False)
    len_colum = temp_df.shape[1]
    temp_df = temp_df.ix[:,[i for i in range(1,len_colum-3)]]
    dropna_df = temp_df.dropna(how='all')
    d_df = d_df.append(dropna_df)

    for i in range(1,len(sheets)):
        temp_df = pd.read_excel(file, sheet_name=i,index=False)
        col_len = temp_df.shape[1]
        temp_df = temp_df.ix[:,[i for i in range(1,col_len-3)]]
        dropna_df = temp_df.dropna(how='all')
        d_df = pd.merge(d_df, dropna_df, how='outer', on=['姓名'])
    # b_df = c_average(file,b_df,len(sheets))
    cls_name = '数学{}'.format(ch)
    d_df[cls_name] = d_df.mean(axis=1)
    d_df = d_df[['姓名',cls_name]]
    d_df = d_df.groupby(['姓名']).max()
    print(d_df.head())
    shuxue_comp = pd.merge(shuxue_comp, d_df, how='outer', on=['姓名'])
    print(shuxue_comp.shape)
```

```python
shuxue_comp['数学平均'] = shuxue_comp.mean(axis=1)
shuxue_comp = shuxue_comp[['姓名','数学平均']]
shuxue_comp.head()
```

```python
comp = zhuanmen_comp
comp = pd.merge(comp, shuxue_comp, how='left', on=['姓名'])
```

```python
comp = comp[['姓名','专业平均','数学平均']]
comp['理解度总平均'] = comp.mean(axis=1)
comp.head()
```
