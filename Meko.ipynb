# 计算学生模拟考平均分，理解度平均值，作业正确率

## 1.需要用的表格
   * 考试成绩/2019冬季  
   * 2019冬季作业登记表
   * 2019冬季课程理解度/
   * 学员调查表part2

```python
import  pandas as pd
import numpy as np 
import os 
from pandas import DataFrame,Series
import re
import xlrd
```

## 2.计算思路

* 模拟考平均分  

```python
df = pd.read_excel("./2019冬季.xlsx",sheetname="情报",header=1)
test_score_df = df[['姓名0','总考分1','总考分2','总考分3']]
test_score_df['考试平均分']=test_score_df.mean(axis=1)
test_score_df.head()
```

```python
test_score_df.shape
```

* 作业正确率

> 合并2019冬季作业登记表当中的情报各班表格

```python
file = '2019冬季作业登记表.xlsx'

# 处理数据，合并多个sheet
wb = xlrd.open_workbook(file)
# 获取全部的工作簿
sheets = wb.sheets()
qingbao_df = DataFrame()
# 从第4个sheet开始合并,到第7个
for i in range(4,7):
    # header=1，那么数据会从第二行开始读取
    temp_df = pd.read_excel(file, sheet_name=i, header=1)
    # how='all'：只要存在数据为空的行都删除，还有其他的方法
    # 每个sheet倒数两行都不要
    dropna_df = temp_df.dropna(how='all')[:-2]
    qingbao_df = qingbao_df.append(dropna_df)
qingbao_df.head()
```

> 取得算平均的区域

```python
qingbao_score = qingbao_df.ix[:,[i for i in range(1,10)]]
qingbao_score.head()
```

> 计算专业课作业平均正确率

```python
qingbao_df["CS_Average"]=qingbao_score.mean(axis=1)
qingbao_average = qingbao_df[['姓名','CS_Average']]
qingbao_average.head()
```

> 计算数学课课作业平均正确率与上面同理

```python
file = '2019冬季作业登记表.xlsx'

# 处理数据，合并多个sheet
wb = xlrd.open_workbook(file)
# 获取全部的工作簿
sheets = wb.sheets()

shuxue_df = DataFrame()
# 从第10个sheet开始合并
for i in range(4):
    # header=1，那么数据会从第二行开始读取
    temp_df = pd.read_excel(file, sheet_name=i, header=1,index=False)
    # how='all'：只要存在数据为空的行都删除，还有其他的方法
    # 每个sheet倒数两行都不要
    dropna_df = temp_df.dropna(how='all')[:-2]
    shuxue_df = shuxue_df.append(dropna_df)
shuxue_df.head()
```

```python
shuxue_score = shuxue_df.ix[:,[i for i in range(2,15)]]
shuxue_df['Math_Average']=shuxue_score.mean(axis=1)
shuxue_average = shuxue_df[['姓名','Math_Average']]
shuxue_average.head()
```

> 专业课作业成绩 left join 数学作业成绩 得出情报专业同学的总作业准确率，再求平均 

```python
homework = pd.merge(qingbao_average, shuxue_average, how='left', on=['姓名'])
homework['Average'] = homework.mean(axis=1)
homework.head()
```
